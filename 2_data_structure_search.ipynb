### Variability of movie ratings

Audiences can have different opinions about a movie. The inconsistency among viewers, or the variability of movie ratings, can be measured using a random variable known as **variance**.

 $variance = \frac{1}{N} \displaystyle\sum_{i=1}^N (x_i - \overline{x})^2 $, where $x_i$  represents an individual movie rating, $N$ is total number of movie rating and $\overline{x}$ is the mean value of the movie ratings.

```python
# {movie_id:movie_title, ...}
movie_title = {m[0]: m[1] for m in movies}

# {movie_id:[movie_rating, ...], ...}
movie_rating = {movie_id:list(ratings[ratings['movieId'] == movie_id]['rating']) for movie_id in np.unique(ratings['movieId'])}

# min number of ratings per movie = 20
movie_rating = {i[0]:i[1] for i in movie_rating.items() if len(i[1]) > 20}

# [(movie_title, variance, num_ratings), ...]
variance_data = sorted([(movie_title[id], np.var(movie_rating[id]), len(movie_rating[id])) for id in movie_rating], key=lambda x: x[1], reverse=True)
variance_data[:3]
```

<a id="h3"></a>
### Variance distribution

Let's see what the distribution of **variance** looks like:

```python
# [variance, ...]
variances = [v[1] for v in variance_data]
```

```python
# histogram
fig, g = plt.subplots(figsize = (21, 9))
g.hist(variances, density=False, bins=50, color='black')
g.set_title('Variance Distribution', loc ='center')
plt.xlabel('X - Variance of movie ratings')
plt.ylabel('Number of filsm')
g.xaxis.set_ticks_position('none')
g.yaxis.set_ticks_position('none')
[g.spines[x].set_visible(False) for x in ['top', 'bottom', 'left', 'right']]
g.grid(visible=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)
plt.show()
```

If we close one eye, we suspect a Normal distribution. The below Quantile-Quantile Plot refutes the hypothesis, as the points are not positioned on the line.

```python
# QQ plot
pp = sm.ProbPlot(np.array(variances), fit=True)
qq = pp.qqplot(marker='.', markerfacecolor='k', markeredgecolor='k', alpha=0.3)
sm.qqline(qq.axes[0], line='45', fmt='k')
plt.grid(visible=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)
plt.xlabel('Teoretični kvantili')
plt.ylabel('Kvantili vzorca')
plt.show()
```

<a id="h4"></a>
### Estimation of distributions parameters

I attempted to estimate the values of the parameters for three known distributions and visualized them.      

```python
# normal distribution
mi = np.mean(variances)
sigma2 = (len(variances) - 1) / len(variances) * np.var(variances)
mi, sigma2
```

```python
# student distribution
student_pars = student.fit(variances)
student_xr = np.linspace(0, 2.2155006, 100)
student_P = [student.pdf(x, *student_pars) for x in student_xr ]
student_pars
```

```python
# beta distribution
beta_pars = beta.fit(variances)
beta_xr = np.linspace(0, 2.2155006, 100)  
beta_P = [beta.pdf(x, *beta_pars) for x in beta_xr ]
beta_pars
```

```python
# histogram
fig, g = plt.subplots(figsize = (21, 9))
counts, bins, _ = g.hist(variances, density=True, label='vzorec', bins=50, color='black')
normal_P = [mvn.pdf(x, mi, sigma2) for x in bins]
g.xaxis.set_ticks_position('none')
g.yaxis.set_ticks_position('none')
[g.spines[x].set_visible(False) for x in ['top', 'bottom', 'left', 'right']]
plt.plot(bins, normal_P, label='Normal', linewidth=2.0, color='purple')
plt.plot(student_xr, student_P, label='Student', linewidth=2.0, color='blue')
plt.plot(beta_xr, beta_P, label='Beta', linewidth=2.0, color='cyan')
plt.grid(visible = True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)
plt.xlabel('X - Varianca ocen filma')
plt.ylabel('P(X)')
plt.legend(loc=1);
plt.show()
```

The above graph confirms that it is a **Beta distribution**, as it best fits the sample.

### Movie clustering

Recommendation systems often discover groups of items (in our case, movies) that exhibit high mutual similarity. I have collected the top 100 most-watched movies and attempted to determine if there are any clusters among them. We can view a movie as a vector, where the number of components is equal to the number of users. The vectors also contain unknown values. The table below illustrates an example of vectors for ten movies.

| x          |Movie                                      | $u_0$ |$u_1$ | $u_2$ | $\cdots$ |
|------------|-------------------------------------------|:-----:|:----:|:------|:--------:|
|$\vec{x_0}$ | Fight Club (1999)                         | ?     | ?    | ?     | $\cdots$ |
|$\vec{x_1}$ | Twelve Monkeys (a.k.a. 12 Monkeys) (1995) | ?     | ?    | 2.5   | $\cdots$ |
|$\vec{x_2}$ | Independence Day (a.k.a. ID4) (1996)      | ?     | ?    | ?     | $\cdots$ |
|$\cdots$    | $\cdots$                                  | $\cdots$ | $\cdots$  | $\cdots$ | $\cdots$ |

```python
# {movie_id, ...}
top_100 = set(dict(sorted(movie_rating.items(), key=lambda x:len(x[1]), reverse=True)[:100]).keys())

# {user_id, ...}
users = {r[0] for r in ratings if r[1] in top_100}

# {(user_id, movie_id):rating, ...}
ratings_opt = {(r[0], r[1]):r[2] for r in ratings if r[1] in top_100}
```

I am storing the movie vectors in the `data` variable as shown in the instructions. Since no similarity measure successfully operated with missing values, I decided to perform **imputation**. I attempted to replace them with:

- *mean value of the attribute*
- *median of the attribute*
- *mean value of movie ratings*
- *fixed label for unknown value*

```python
# [[rating1, rating2, ..., movie_id]]
data = [[ratings_opt[u, m] if (u, m) in ratings_opt else np.nan for u in users] + [m] for m in top_100]

# imputation
for i in range(0, len(data)):
    for j in range(0, len(data[i])):
        if np.isnan(data[i][j]):
            #data[i][j] = np.nanmean([d[j] for d in data])                     mean value of the attribute
            #data[i][j] = np.nanmedian([d[j] for d in data])                   median of the attribute
            #data[i][j] = np.average(movie_rating[data[i][len(data[i]) - 1]])  mean value of movie ratings
            data[i][j] = -1                                                  # fixed label for unknown value
```

None of the above-mentioned methods nicely separated the movies into clusters. Satisfactory clustering was only achieved by using *a fixed label for unknown values*, which in my case is an unreachable rating value **-1**.

For the clustering algorithm, I chose **Hierarchical Clustering** due to the large number of attributes and data in the problem, as well as the fact that I don't have a predetermined number of clusters I want to create. For the similarity measure, I selected the **Jaccard index**, which was the only one that effectively displayed clusters from the data.

```python
# similarity measure
L = sch.linkage(data, method="average", metric="jaccard")
t = 0.3
predictions = sch.fcluster(L, t=t, criterion="distance").ravel()
score = silhouette_score(data, predictions, metric="euclidean")
score
```

```python
# [Orange.data.ContinuousVariable(atribut1), ..., Or.data.ContinuousVariable('movie')]
attributes = [Or.data.ContinuousVariable(u) for u in users] + [Or.data.ContinuousVariable('movie')] 

# orange table
data = Or.data.Table.from_list(Or.data.Domain(attributes), data)
```

```python
# dendrogram
plt.figure(figsize=(25, 15))
labels = [movie_title[row["movie"].value] for row in data]
D = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Distance")
plt.show()
```

In my opinion, we could achieve much better results with differently shaped attributes. Nevertheless, on the above dendrogram, we can see 4 clusters of different colors:
