### Data preparation

Project uses data from database for Hollywood movies called [MovieLens](https://grouplens.org/datasets/movielens/) covering the period 1995-2016 for the purpose of rating recommendation systems. It contains information about viewers and their ratings for individual films on a scale of 1 to 5. In addition to the basic user rating matrix, it includes additional data about the movies such as genre, date, tags, and cast.

The data is located in the folder `./data/ml-latest-small`. The dataset consists of the following files:

* `tings.csv`: user and rating data,
* `movies.csv`: movie genre data,
* `cast.csv`: actor data,
* `tags.csv`: tag data,
* `links.csv`: links to related datasets.

You can find more details about the dataset on the [spletni strani](http://files.grouplens.org/datasets/movielens/ml-latest-small-README.html).

```python
import numpy as np
import math
import warnings
from csv import DictReader
import matplotlib.pyplot as plt
from collections import Counter
from datetime import datetime as dt
```

```python
reader = DictReader(open('data/ml-latest-small/movies.csv', 'rt', encoding='utf-8'))
movies = [[int(row["movieId"]), row["title"], set(row["genres"].split("|"))] for row in reader]
```

```python
ratings = np.loadtxt('data/ml-latest-small/ratings.csv', delimiter=",", skiprows=1)
```

```python
reader = DictReader(open('data/ml-latest-small/cast.csv', 'rt', encoding='utf-8'))
cast = [[int(row["movieId"]), set(row["cast"].split("|"))] for row in reader]
```

<a id="h2"></a>
### Which is all-time best movie? 

The best movie is the one that has **the highest ratio of votes to average rating**. The descending sorted list *real_ratings* will contain the actually highest-rated films at the top (Forrest Gump, Jurassic Park, Silence of the Lambs).

$$best\_ movie_i = max(\frac{count(votes_i)}{avg(ratings_i)})$$


```python
warnings.filterwarnings(action='ignore', message='Mean of empty slice')

# {movie_id:num_votes}
votes = Counter(int(r[1]) for r in ratings)

# {movie_id:avg_rating}
avgs = {m[0]:np.nanmean(ratings[(ratings[:, 1] == m[0], 2)]) for m in movies}
```

```python
# [(ratio_votes_rating, num_votes, avg_rating, movie_name)]
real_ratings = [(votes[m[0]]/avgs[m[0]], votes[m[0]], avgs[m[0]], m[1]) for m in movies]
real_ratings = sorted([r for r in real_ratings if not math.isnan(r[0])], reverse=True)
real_ratings[:3]
```

### Genre distribution

It is interesting to depict the distribution of genres in the film industry, as we discover that the majority of films fall into the **drama** category, followed by **comedies**. This is somewhat trivial, as they encompass the widest audience base.

```python
# {genre}
genres = set().union(*[m[2] for m in movies])

# [genre]
all_genres = [genre for m in movies for genre in m[2]]

# {genre:num_movies}
genres_counter = dict(Counter(all_genres).most_common())
```

```python
# horizontal bar chart
fig, g = plt.subplots(figsize = (21, 9))
g.barh(list(genres_counter), list(genres_counter.values()), color="black")
g.set_title('Genre Distribution', loc ='center')
plt.xlabel("Number of movies")
plt.ylabel("Genre")
g.xaxis.set_ticks_position('none')
g.yaxis.set_ticks_position('none')
[g.spines[x].set_visible(False) for x in ["top", "bottom", "left", "right"]]
g.grid(visible = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
g.invert_yaxis()
for i in g.patches:
    plt.text(i.get_width() + 20, i.get_y() + 0.5, 
             str(round((i.get_width()), 2)), 
             fontsize = 10, fontweight ='bold', color ='black')
plt.show()
```

### Is there a correlation between watch-time and the average rating of a film?

To calculate the correlation between watch-time and the average rating of a film, I used the Pearson correlation coefficient, which can be easily computed using the `corrcoef` function from the `numpy` library. It takes as input the list of view counts *num_views* and the list of average ratings *avg_ratings*. The coefficient is calculated using the formula:

$$\rho = \frac{(x - \bar{x})(y - \bar{y})}{\sigma_x \sigma_y}$$

```python
# [num_views]
num_views = list(zip(*real_ratings))[1]

# [rating]
avg_ratings = list(zip(*real_ratings))[2]
```

```python
# pearson correlation
np.corrcoef(avg_ratings, num_views)[1,0]
```

The resulting coefficient ranges from -1 to 1, where a value of -1 indicates an inverse correlation, a value of 1 indicates a direct correlation, and a value close to 0 suggests little to no correlation between the random variables $X$ and $Y$. The above result suggests that the variables are relatively independent or **weakly correlated**, which is further confirmed by the scatter plot below, where most points are not linearly distributed.

```python
# scatter plot
fig, g = plt.subplots(figsize = (21, 9))
g.plot(avg_ratings, num_views, "k.")
g.set_title('Correlation', loc ='center')
plt.ylabel("Number of Views")
plt.xlabel("Average Rating")
g.xaxis.set_ticks_position('none')
g.yaxis.set_ticks_position('none')
g.grid(visible = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
[g.spines[x].set_visible(False) for x in ["top", "bottom", "left", "right"]]
plt.show()
```

### Movie popularity over time

In the following graphs, we can observe how the popularity of movies changes over time due to various factors. The popularity of the movie **Toy Story** significantly increased after the release of its sequel, Toy Story 3, in 2010, while the film **The Lion King** became more popular after the premiere of its IMAX version in 2002.

```python
# [(date, ratings)]
film = [(dt.fromtimestamp(int(r[3])).strftime('%Y-%m-%d %H:%M'), int(r[2])) for r in ratings if int(r[1]) == 1]
film2 = [(dt.fromtimestamp(int(r[3])).strftime('%Y-%m-%d %H:%M'), int(r[2])) for r in ratings if int(r[1]) == 364]
film.sort()
film2.sort()

# [(date, avg_last_30_ratings)]
film_avgs = [(film[i][0], sum(map(lambda x: x[1], film[i - 29:i + 1])) / 30) for i in range(29, len(film))]
film_avgs2 = [(film2[i][0], sum(map(lambda x: x[1], film2[i - 29:i + 1])) / 30) for i in range(29, len(film2))]

# [date]
dates = list(zip(*film_avgs))[0]
dates2 = list(zip(*film_avgs2))[0]

# [avg_last_30_ratings]
avg_ratings = list(zip(*film_avgs))[1]
avg_ratings2 = list(zip(*film_avgs2))[1]
```

```python
fig, (g1, g2) = plt.subplots(1, 2, figsize = (21, 9))

# Toy Story (1995)
g1.plot(dates, avg_ratings, '-', color="black")
g1.set_title('Popularity of Toy Story (1995)', loc ='center')
g1.xaxis.set_ticks_position('none')
g1.yaxis.set_ticks_position('none')
g1.grid(visible = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
[g1.spines[x].set_visible(False) for x in ["top", "bottom", "left", "right"]]
plt.sca(g1)
plt.ylabel("Average Rating")
plt.xlabel("Time")
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
plt.xticks(dates[::int(len(dates) / 4)])

# Lion King (1994)
g2.plot(dates2, avg_ratings2, '-', color="black")
g2.set_title('Popularity of The Lion King (1994)', loc ='center')
g2.xaxis.set_ticks_position('none')
g2.yaxis.set_ticks_position('none')
g2.grid(visible = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
[g2.spines[x].set_visible(False) for x in ["top", "bottom", "left", "right"]]
plt.sca(g2)
plt.ylabel("Average Rating")
plt.xlabel("Time")
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
plt.xticks(dates2[::int(len(dates2) / 4)])
plt.show()
```

### What is popularity anyway?

From **ChatGPT**: "*Popularity refers to the state or condition of being widely liked, admired, or recognized by a large number of people.*" In my case popularity is calcualated by summing up the *real_rating* of movies in which an actor has appeared. In this assessment, I take into account the number of movies the actor has played in (more appearances earn more points) and the actual quality of their acting (based on viewer ratings).

$$popularity_{actor} = \sum_{k = 1}^{n} real\_rating_{i,k} = \sum_{k = 1}^{n} \frac{count(ratings\_movie_i}{avg(ratings\_movie_i)}$$

```python
# {movie_id:ratio_votes_rating}
real_ratings_dict = {m[0]:votes[m[0]]/avgs[m[0]] if not math.isnan(votes[m[0]]/avgs[m[0]]) else 0 for m in movies}

# {actor:popularity}
actors = {a: 0 for a in set().union(*[m[1] for m in cast]) if a != ""}

#  if an actor appears in a movie, their real_rating for that film is added
for c in cast:
    for a in c[1]:
        if a != "":
            actors[a] += real_ratings_dict[c[0]]

# top 3 most popular actors
actors = sorted(actors.items(), key=lambda kv: kv[1], reverse=True)
actors[:3]
```
